```python
df = pd.read_parquet(ruta_file)
df["utes_length"] = df.utes.apply(len)
df_con_utes = df[df.utes_length>0]
df_con_utes
```

```python
# Filtrar las filas donde 'integrantes' es NaN
filas_con_nan = df_con_utes[df_con_utes['utes'].isna()]

# Calcular el total de filas con 'integrantes' a NaN
total_nan = filas_con_nan.shape[0]
print(total_nan)
```

```python
# Coger lista de compañías con utes
all_comp = df_con_utes.FullName.values.tolist()
all_utes = df_con_utes.utes.values.tolist()

#Para aplanar la lista
def flatten(xss):
    return [x for xs in xss for x in xs]

#El set para eliminar duplicados 
all_utes_d = list(set(flatten(all_utes)))

#print(f"El total de utes es: {len(all_utes_d)}\n")
#print(f"El total de compañías es: {len(all_comp)}\n")
#print(f"all_comp: {all_comp[0:10]}\n")
#print(f"all_utes: {all_utes[0:10]}\n")
#print(f"all_utes_d: {all_utes_d[0:10]}\n")
```

```python
utes_dict = {
    ute: [] for ute in all_utes_d
}
```

```python
keys = len(utes_dict)
print(f"El número de keys en el diccionario es: {keys}")
```

```python
for comp,utes in zip(all_comp,all_utes):
    for ute in utes:
        utes_dict[ute].append(comp)
```
